```python
import json
import os
from PIL import Image
```

```python
# Путь к папке с изображениями и JSON файлами
base_dir = "C://ml3"
train_images_dir = os.path.join(base_dir, 'train_images')
test_images_dir = os.path.join(base_dir, 'test_images')

# Путь к JSON файлам
train_annotations_path = os.path.join(base_dir, 'usdc_train.json')
test_annotations_path = os.path.join(base_dir, 'test_file_names.json')
```

```python
# Загрузка аннотаций для обучающего набора
with open(train_annotations_path, 'r') as train_file:
    train_data = json.load(train_file)

# Загрузка аннотаций для тестового набора
with open(test_annotations_path, 'r') as test_file:
    test_data = json.load(test_file)
```

```python
# Предполагаем, что в данных есть ключи 'images' и 'annotations'
train_images = train_data['images']
train_annotations = train_data['annotations']

test_images = test_data['images']
```

```python
def load_image(image_name, directory):
    image_path = os.path.join(directory, image_name)
    return Image.open(image_path)

# Пример загрузки изображений из обучающего набора
for image_info in train_images:
    image_name = image_info['file_name']  # Предполагаем, что имя файла хранится здесь
    image = load_image(image_name, train_images_dir)
    # Здесь вы можете обработать изображение или вывести его
    print(f"Загружено изображение: {image_name}")

# Пример загрузки изображений из тестового набора
for image_info in test_images:
    image_name = image_info['file_name']
    image = load_image(image_name, test_images_dir)
    print(f"Загружено тестовое изображение: {image_name}")
```
